```python
import numpy as np
import matplotlib.pyplot as plt

conv_coarse = np.loadtxt("../POD/case1/coarse/convergence.txt")
x_conv = np.arange(len(conv_coarse))
conv_medium = np.loadtxt("../POD/case1/medium/convergence.txt")
x_med = np.arange(len(conv_medium))
conv_fine = np.loadtxt("../POD/case1/fine/convergence.txt")
x_fine = np.arange(len(conv_fine))

plt.figure()
plt.plot(x_conv, conv_coarse, label = 'coarse')
plt.plot(x_med, conv_medium, label = 'medium')
plt.plot(x_fine, conv_fine, label = 'fine')
plt.legend()
plt.yscale("log")
```
